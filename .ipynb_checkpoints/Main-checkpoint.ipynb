<p><h4>Precleaning the data</h4></p>
Converting the format and data type of "price" column, and then clean(pre_process) the dataframe.

For numerical variables, NaN value is imputed by the mean. Simply dropping the entire row will remove information that may be useful for our prediction. Because there's no abnormal high value, mean value can be the best one to be imputed.

For categorical variables, we encode them into dummies by removing original columns with data type of "object" and concatenate dummy columns into the dataframe. This dummy columns can help us demonstrate which category in which variable more contributes to the target variable.

```python
def pre_process(df, X_columns, y_column):
    '''
    INPUT:
    df - the dataframe 
    
    OUTPUT:
    X - cleaned dataframe that containes all variables which may affect target variable
    y - cleaned dataframe of target variable
    
    Split the original dataframe to target variable and others. Clean the data by
    1) Drop the row with NaN target variable
    2) Imputing NaN values for numerical variables
    3) Create dummies for categorical values
    '''
    
    df[y_column].replace("[\$,]", "", regex=True, inplace=True)
    df[y_column] = df[y_column].astype(float)
    
    df = df.dropna(subset=[y_column], axis=0)
    
    y = df[[y_column]]
    
    X = df[]
    
    num_vars = X.select_dtypes(include=["int", "float"]).columns
    
    for col in num_vars:
        X[col].fillna((X[col].mean()), inplace=True)
    
    cat_vars = X.select_dtypes(include=["object"]).columns
    
    for var in cat_vars:
        X = pd.concat([X.drop([var], axis=1), pd.get_dummies(X[var], prefix=var, prefix_sep='_', drop_first=True)], axis=1)
    
    return X, y
```

```python
X_columns = ["neighbourhood_cleansed", "property_type", "room_type", "accommodates", "bathrooms", "bedrooms", "beds", "bed_type"]
X_seattle, y_seattle = pre_process(seattle_listings, X_columns, "price")
X_seattle.head()
```

<p><h4>Building model</h4></p>

```python
#Split into train and test
X_sea_train, X_sea_test, y_sea_train, y_sea_test = train_test_split(X_seattle, y_seattle, test_size = .30, random_state=42) 

lm_model_sea = LinearRegression(normalize=True) # Instantiate
lm_model_sea.fit(X_sea_train, y_sea_train) # Fit
        
#Predict and score the model
y_sea_test_preds = lm_model_sea.predict(X_sea_test) 
"The r-squared score for your model was {} on {} values.".format(r2_score(y_sea_test, y_sea_test_preds), len(y_sea_test))
```

```python
X_sea_train.shape
```

```python
lm_model_sea.coef_.shape
```

```python
def coef_weights(coefficients, X_train):
    '''
    INPUT:
    coefficients - the coefficients of the linear model 
    X_train - the training data, so the column names can be used
    OUTPUT:
    coefs_df - a dataframe holding the coefficient, estimate, and abs(estimate)
    
    Provides a dataframe that can be used to understand the most influential coefficients
    in a linear model by providing the coefficient estimates along with the name of the 
    variable attached to the coefficient.
    '''
    coefs_df = pd.DataFrame()
    coefs_df['est_int'] = X_train.columns
    coefs_df['coefs'] = coefficients
    coefs_df['abs_coefs'] = np.abs(coefficients)
    coefs_df = coefs_df.sort_values('abs_coefs', ascending=False)
    return coefs_df
```

```python
coef_sea_df = coef_weights(lm_model_sea.coef_[0], X_sea_train)
coef_sea_df.head(30)
```

<h4>Conclusions</h4>
<p>In conlusion, the location of the Airbnb property plays the most important role in its pricing , both in Seattle and Boston. Some locations are not ideal for Airbnb business.

We can see from the chart that in Seattle, price drops significantly as compared to Boston.

Property type can also affect the price. Some rare property types like 'Boat' and 'Dorm' in Seattle are top indicators of the pricing, while in Boston it does not have greater effect. Whereas the lack of data may leads to inappropriate judgements on how much the price indicated by property type. In terms of room type, sharing room usually indicates a low price both in Seattle and Boston. The number of bedrooms, beds, bathrooms and bed type is also an indicator, but not as crucial as location and property/room type.</p>

```python
def func(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:.1f}%".format(pct)
```
